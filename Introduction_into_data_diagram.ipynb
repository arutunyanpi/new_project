```python
import pandas as pd
url = 'https://raw.githubusercontent.com/Sketchjar/MachineLearningHD/main/bank_marketing_dataset.csv'
WorkDataFrame=pd.read_csv(url)
different_price=set(WorkDataFrame['cons.price.idx'])
WorkDataFrame.head()
```

```python
max_age_in_job=WorkDataFrame.groupby('job')['age'].max().reset_index()
mean_price_c=WorkDataFrame.groupby('job')['cons.price.idx'].mean().reset_index()
```

```python
mean_price_c
```

```python
max_age_in_job.columns=['Отрасль','Максимальный_возраст']
mean_price_c.columns=['Job','Mean_price']
max_age_in_job
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))

plt.title('Cредний возраст по отраслям')
plt.xlabel('Отрасль')
plt.ylabel('Максимальный возраст')

plt.ylim(0,90)

plt.grid() #Отрисовка сетки 

plt.scatter(
    max_age_in_job['Отрасль'],
    max_age_in_job['Максимальный_возраст'],
    alpha=0.7, #прозрачность точек
    color = '#000'
)
plt.scatter(
    mean_price_c.Job,
    mean_price_c.Mean_price,
    alpha=0.7, #прозрачность точек
    color = 'pink'
)
plt.show() #Отрисовка текущего графика

```
